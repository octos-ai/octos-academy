### Step 13. Turn the item price into a float

#### Step 13.a. Check the item price type

```python
chipo['item_price']

#Poderia ter usado chipo.info
#Ao usar type(chipo['item_price'][0]), tem como retorno tipo string
```

#### Step 13.b. Create a lambda function and change the type of item price

```python
#Troca a coluna original por uma cópia, que não tem o cifrão nas entradas
#Astype vai transformar essas entradas no tipo float (com cifrão não era possível)
#Se colocar x.replace em vez de str(x).replace, só dá certo uma vez, porque float não tem replace

muda_tipo = lambda y: y.apply(lambda x: str(x).replace('$','')).astype('float64')
chipo['item_price'] = muda_tipo(chipo['item_price'])
```

### Step 14. How much was the revenue for the period in the dataset?

```python
#Para cada índice, multiplicar a entrada do preço com a entrada da quantidade
#Somar tudo no final

sales = chipo['quantity'].mul(chipo['item_price'])
revenue = sum(sales)
revenue
```
