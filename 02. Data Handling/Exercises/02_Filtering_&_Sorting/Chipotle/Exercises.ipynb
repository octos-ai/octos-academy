### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_csv(url , sep = '\t')
```

### Step 4. How many products cost more than $10.00?

```python
#Primeiro temos que observar que o tipo das informações contidas na coluna item_price é object
#Para trabalhar com as informações dessa coluna corretamente temos que convertê-la para o tipo float w
chipo['item_price'].dtype
```

```python
#Usando uma função lambda para substituir o $ por um espaço vazio
chipo['item_price'] = chipo['item_price'].apply(lambda x: str(x).replace("$" , " "))
```

```python
#Convertendo a coluna item_price para float
chipo['item_price'] = chipo['item_price'].astype('float64')
```

```python
#Observamos se a coluna foi convertida com sucesso
chipo['item_price'].dtype
```

```python
#Remover as repetições com o objetivo de contar apenas um único produto de cada
#Afunção .drop_duplicates - de acordo com a documentação do pandas - retornar o DataFrame removendo as linhas duplicadas
remove_repeticoes = chipo.drop_duplicates(['item_name' , 'quantity'])
```

```python
#Utilizando dados com uma quantidade
#Nesses dados temos no item_price o preço de um único produto
produto_unico = remove_repeticoes[remove_repeticoes.quantity == 1]
```

```python
#Solicitando quantos produtos custam mais de $10.00
#A função .nunique() - de acordo com a documentação do pandas - conta observações distintas sobre o eixo solicitado
produto_unico[produto_unico['item_price'] > 10].item_name.nunique()
```

### Step 5. What is the price of each item?
